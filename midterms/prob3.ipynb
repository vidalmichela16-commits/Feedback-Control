```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients of the transfer function G(s)
# Numerator: 3
# Denominator: 8s^4 + 12s^3 + 26s^2 + 18s + 0
num = [3]
den = [8, 12, 26, 18, 0]

# Define the Transfer Function using the control library
G = ctrl.TransferFunction(num, den)

# 1. Print the Transfer Function properly aligned
print("Transfer Function G(s):")
print(G)
```

```python
# 2. Print the values of the poles and zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
# 3. Generate Step Response plot
plt.figure(figsize=(10, 6), dpi=100)
t, y = ctrl.step_response(G)
plt.plot(t, y, color='#0D9488', linewidth=2)
plt.title('System Step Response Analysis', fontsize=14, fontweight='bold')
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Amplitude X3(t)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 4. Visualize the poles and zeros on the complex plane
plt.figure(figsize=(10, 6), dpi=100)
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map of G(s)', fontsize=14, fontweight='bold')
plt.show()
```
