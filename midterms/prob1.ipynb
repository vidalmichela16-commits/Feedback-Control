```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the Transfer Function
# Numerator: s + 1 -> coefficients [1, 1]
# Denominator: s^4 + 4s^3 + 7s^2 + 6s + 3 -> coefficients [1, 4, 7, 6, 3]
num = [1, 1]
den = [1, 4, 7, 6, 3]

G = ctrl.TransferFunction(num, den)

# Display the transfer function
print("Transfer Function:")
print(f"G(s) = {G}")
```

```python
# Step 2: Calculate and print Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)

print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
# Step 3: Generate Step Response Plot
t, y = ctrl.step_response(G)

plt.figure(figsize=(10, 5))
plt.plot(t, y, color='#0D9488', linewidth=2)
plt.title('Step Response of the System', fontsize=14, fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Step 4: Generate Pole-Zero Map
plt.figure(figsize=(6, 6))
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map', fontsize=14, fontweight='bold')
plt.show()
```
